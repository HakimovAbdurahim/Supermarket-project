```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline
```

```python
df = pd.read_csv('sotuvlar.csv')
df['Date'] = df['Date'].astype(str)
df["Date"] = pd.to_datetime(df["Date"],infer_datetime_format=True)
df['Time'] = pd.to_datetime(df['Time'])
df['Hour'] = (df['Time']).dt.hour
df['Hour'].unique()
```

```python
count_gender = df.groupby(['City','Gender']).count()['Customer type']
count_gender.unstack(level=0).plot(figsize=(10,6),kind='bar')
plt.xlabel('Jinslar turi')
plt.ylabel('Soni')
plt.title('Har bir shaharda erkaklar va ayollar')
plt.legend(loc='upper center')
```

```python
customer_type = df.groupby(['City','Customer type']).count()['Invoice ID']
customer_type.unstack(level=0).plot(figsize=(10,6),kind='bar', color='rgb')
plt.xlabel('Haridorlar')
plt.ylabel('soni')
plt.title('Supermarketning haridorlar turi')
plt.legend(loc='lower center')
```

```python
count_gender = df.groupby(['Gender','Product line']).count()['Invoice ID']
count_gender.unstack(level=0).plot(figsize=(10,7),kind='bar', color='rgb')
plt.title('Har bir mahsulot uchun bolgan talab')
plt.xlabel('Mahsulotlar')
plt.ylabel('Miqdori')
plt.legend(loc='lower right')
```

```python
normal_member = df.groupby(['Customer type','Product line']).count()['Gender']
normal_member.unstack(level=0).plot(figsize=(10,7),kind='bar', color=['orange','red'])
plt.title('Mahsulotlarni sotib olish boyicha Member va Normal ortasidagi farq')
plt.xlabel('Mahsulotlar')
plt.ylabel('Miqdori')
plt.legend(loc='lower right')
```

```python
count_product = df.groupby(['City','Product line']).count()['Invoice ID']
count_product.unstack(level=0).plot(figsize=(10,7),kind='bar', color='rgb')
plt.xlabel('Mahsulotlar')
plt.ylabel('Miqdori')
plt.title('3ta shahar boyicha sotib olingan mahsulotlar va ularning miqdori')
plt.legend(loc='lower right')
```

```python
count_rating = df.groupby(['Product line']).count()['Rating']
count_rating.plot(figsize=(8,6),kind='barh', color='green')
plt.xlabel('Rating')
plt.ylabel('Mahsulotlar')
plt.title('Mahsulotlarning umumiy Rating korsatkichlari')
```

```python
colors = ['green','blue','red']
payment = df.groupby(['Payment']).count()['Invoice ID']
payment.plot(figsize=(16,6),kind='pie', autopct='%1.1f%%',
             colors=colors, shadow=True,explode = (0.05, 0.05, 0.05))
plt.ylabel('')
plt.xlabel('Tolov turi va umumiy korinishi')
plt.legend(loc='lower right')
```

```python
sum_gross = df.groupby(['City'])['gross income'].sum()
sum_gross.plot(figsize=(16,6),kind='pie', autopct='%1.1f%%', colors=colors,
               shadow=True, explode=(0.03, 0.05, 0.03))
plt.ylabel('')
plt.title('Shaharlarning 3 oylik yalpi daromad korsatkichi')
plt.legend(loc='lower left')
```
